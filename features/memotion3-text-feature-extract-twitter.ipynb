```python
from transformers import AutoTokenizer, AutoModel, TFAutoModel
import numpy as np

MODEL = "cardiffnlp/twitter-roberta-base"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
```

```python
import pandas as pd
train_image_path = '../input/memotion3/trainImages/trainImages'
val_image_path = '../input/memotion3/valImages/valImages'

train_data = pd.read_csv('../input/memotion3/train.csv')
val_data = pd.read_csv('../input/memotion3/val.csv',sep ='\t')#,nrows=10
test_data = pd.read_csv('../input/memotion3/test.csv')
```

```python
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)
```

```python
data = np.array(test_data)
texts, images, labels = [], [], []
idxs = list(range(len(data)))
#             np.random.shuffle(idxs)
    
for i in idxs:
#                 print(i)
    d = data[i]
        # print(d)
    text = preprocess(d[2])
#     print(text)
    encoded_input = tokenizer(text, return_tensors='tf')
    model = TFAutoModel.from_pretrained(MODEL)
    features = model(encoded_input)
    features = features[0].numpy()
    features_mean = np.mean(features[0], axis=0)
                
    # label = d[2]
                
    texts.append(features_mean)
#       images.append(img_array)
    # labels.append(label)
    
        # f.write(features_mean)
    # return np.array(texts), np.array(labels)
pd.DataFrame(texts).to_csv('test_text_feature.csv',index=None)
```
